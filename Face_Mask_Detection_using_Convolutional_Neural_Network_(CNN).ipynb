# Face Mask Detection using Convolutional Neural Network (CNN)

## Project Overview

This project aims to build a Deep Learning model using Convolutional Neural Networks (CNN) to detect whether a person is wearing a face mask or not.
The model is trained on image data and can also perform real-time detection using a live camera.

## Problem Statement

During the COVID-19 pandemic, wearing a face mask became essential to prevent the spread of the virus. However,
monitoring mask compliance manually in public places is difficult.

The objective of this project is to develop a Deep Learning-based Face Mask Detection system that can automatically
classify whether a person is wearing a mask or not using image data and live video feed.

## Project Objectives

- To collect and preprocess image data for mask and no-mask categories.
- To build a CNN model for binary image classification.
- To train and evaluate the model for accuracy and performance.
- To implement real-time face mask detection using webcam.
- To analyze model performance using accuracy and loss metrics.

In this section, we import all the necessary libraries required for image processing, model building, and visualization.

```python
path1=r"C:\Neural Networks\withmask_withoutmask"
cate=['with_mask','without_mask']
# create the path
```

## Data Loading and Preprocessing

In this section, we load the dataset, resize images, convert them into arrays, and normalize pixel values.

```python
image_size=200
input_image=[]
for i in cate:
    folders=os.path.join(path1,i)
    label=cate.index(i)
    for image in os.listdir(folders):
        image_path=os.path.join(folders,image)
        image_array=cv2.imread(image_path)
        image_array=cv2.resize(image_array,(image_size,image_size))
        input_image.append([image_array,label])
```

### Explanation

- Images are resized to a fixed size for uniform input.
- Pixel values are normalized (0вЂ“255 в†’ 0вЂ“1) to improve model performance.
- Labels are assigned: 
  - 1 в†’ Without Mask
  - 0 в†’ With Mask

```python
len(input_image)
```

```python
np.random.shuffle(input_image)
```

```python
train  = input_image[0:6000]
test   = input_image[6000::] 

X_train = []
Y_train = []

for X_values , labels in train: 

    X_train.append(X_values)

    Y_train.append(labels)

X_test = []

Y_test = []

for X_values , labels in train: 

    X_test.append(X_values)

    Y_test.append(labels)
```

```python
X_train = np.array(X_train)

Y_train= np.array(Y_train)

X_test = np.array(X_test)

#Y_test= np.array(Y_test) # directly used in confusion matrix so no need to convert
```

```python
X_train  = X_train / 255 

X_test  = X_test / 255
```

## Building the CNN Model

We create a Convolutional Neural Network model consisting of:
- Convolution layers for feature extraction
- MaxPooling layers for dimensionality reduction
- Dense layers for classification

```python
import tensorflow as tf
import keras
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense
```

```python
model = Sequential()
model.add(Conv2D(filters = 32 , kernel_size =(5,5) , activation ='relu'  ,padding='same'   ))
model.add(MaxPool2D(pool_size =(2,2)))
model.add(Flatten())
model.add(Dense(256 , activation = 'relu' , input_shape = X_train.shape[1:]))
model.add(Dense(2 , activation= 'softmax'))
```

### Explanation of CNN Layers

- **Conv2D** в†’ Extracts important features from images.
- **MaxPooling2D** в†’ Reduces spatial dimensions.
- **Flatten** в†’ Converts 2D feature maps into 1D vector.
- **Dense** в†’ Fully connected layer for classification.
- **Dropout** в†’ Prevents overfitting.

## Model Training

The model is trained using training data and validated using validation data.

```python
adam=tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=adam,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=5,batch_size=64,validation_split=0.2)
```

### Explanation

- The model learns patterns from images during training.
- Accuracy and loss are tracked for both training and validation datasets.
- Epochs determine how many times the model sees the entire dataset.

```python
pred=model.predict(X_test)
```

```python
pred_cate=np.argmax(pred,axis=1)
```

## Model Evaluation

The trained model is evaluated on test data to measure its performance.

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(Y_test,pred_cate))
print("-----------------------------------------------------")
print(classification_report(Y_test,pred_cate))
```

### Evaluation Metrics

- **Accuracy** в†’ Percentage of correct predictions.
- **Loss** в†’ Error between predicted and actual values.

## Real-Time Face Mask Detection (Live Camera)

Using OpenCV, we access the webcam and detect faces in real-time. The trained model predicts whether the detected face is wearing a mask or not.

```python
import cv2
import numpy as np
from PIL import Image
from keras.models import load_model  #  used to load the pre trained model
```

```python
def  face_extractor(img):

    faces = face_cascade.detectMultiScale(img, scaleFactor = 1.5 , minNeighbors = 5)

    if faces is ():

        return None

    

    

    for (x ,y , w, h)  in faces:

        cv2.rectangle(img , (x,y) , (x+w , y+h), (0,0, 255) , 2)

        roi =  img[y :y+h , x:x+w]

    

    return roi
```

### Explanation

- Webcam captures live video frames.
- Face detection is performed using Haar Cascade.
- Each detected face is passed to the CNN model.
- Prediction is displayed on the screen.

```python
face_cascade =  cv2.CascadeClassifier(r"C:\Users\Chinnarajan M\Downloads\haarcascade_frontalface_default.xml")

model = load_model("withmask_withoutmask.h5")  # name of pre trained model

 

video_capture =cv2.VideoCapture(0) #  camera 

while True:

    ret,frame = video_capture.read()

    

    face = face_extractor(frame)

    if type(face) is np.ndarray:

        face = cv2.resize(face , (200,200))

        im = Image.fromarray(face , 'RGB')

        

        

        img_array = np.array(im)

        

        img_array = np.expand_dims(img_array ,  axis = 0)

        pred = model.predict(img_array)  #  making the prediction

        print(pred)

        

        

        name = "None Matching"

 

        if pred[0][0]  > .5:

            name ="Masked"

        elif pred[0][1]  > .5:

            name = 'Mask is not there , Put on the mask' 

        cv2.putText(frame ,name , (75 , 75) , cv2.FONT_HERSHEY_COMPLEX , 1 , (255 , 0, 0) , 2)    

    else:

        cv2.putText(frame ,"Face not found" , (75 , 75 ) , cv2.FONT_HERSHEY_COMPLEX ,1 , (255 , 0, 0) , 2)

    cv2.imshow('Video' , frame)    

        

    if cv2.waitKey(20) & 0xff == ord('q') : 

        break

     

video_capture.release()

cv2.destroyAllWindows()
```
